```python

import pandas as pd
import matplotlib.pyplot as plt

def load_data(file_path):
    return pd.read_csv(file_path)

def clean_data(df):
    df.drop_duplicates(inplace=True)
    df.fillna(0, inplace=True)
    return df

def add_revenue(df):
    df["Revenue"] = df["Quantity"] * df["Price"]
    return df

def main():
    df = load_data("sales.csv")
    df = clean_data(df)
    df = add_revenue(df)

    print("Total Revenue:", df["Revenue"].sum())
    print("Top Product:", df.groupby("Product")["Revenue"].sum().idxmax())

    summary = df.groupby("Product")["Revenue"].sum()
    summary.plot(kind="bar")
    plt.show()

if __name__ == "__main__":
    main()


```

```python
#description 
# a python based sales data analysis project that loads csv data,cleans it,calculates total revenue,identifies the top product,and visualizes results using bar chart.

#conclusion
# efficient revenue analysis and visualization using python.
```

```python
# sales_analysis.py

import pandas as pd
import sys
import os

def create_sample_file(file_path):
    data = {
        "Product": ["Laptop", "Mobile", "Mouse", "Laptop", "Mobile"],
        "Quantity": [2, 5, 10, 1, 3],
        "Price": [50000, 20000, 500, 50000, 20000]
    }
    df = pd.DataFrame(data)
    df.to_csv(file_path, index=False)
    print("Sample sales.csv file created.\n")

def analyze_sales(file_path):
    df = pd.read_csv(file_path)

    # Data Cleaning
    df.drop_duplicates(inplace=True)
    df.fillna(0, inplace=True)

    # Revenue column
    df["Revenue"] = df["Quantity"] * df["Price"]

    # Calculations
    total_revenue = df["Revenue"].sum()
    total_quantity = df["Quantity"].sum()
    average_sale = df["Revenue"].mean()
    top_product = df.groupby("Product")["Revenue"].sum().idxmax()

    # Report
    print("========= SALES REPORT =========")
    print(f"Total Revenue     : {total_revenue}")
    print(f"Total Quantity    : {total_quantity}")
    print(f"Average Sale      : {average_sale}")
    print(f"Top Selling Prod  : {top_product}")
    print("================================")

def main():
    file_path = "sales.csv"

    # Agar file exist nahi karti to sample bana do
    if not os.path.exists(file_path):
        create_sample_file(file_path)

    analyze_sales(file_path)

if __name__ == "__main__":
    main()
```

```python
#description
#this project performs sales data analysis using python and pandas. It creates a sample dataset,cleans the data,calculates total revenue,and identifies the top selling product.

#conclusin
#simple and efficient sales analysis using python.
```
