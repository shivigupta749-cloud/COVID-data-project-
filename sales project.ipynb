```python

import pandas as pd
import matplotlib.pyplot as plt

def load_data(file_path):
    return pd.read_csv(file_path)

def clean_data(df):
    df.drop_duplicates(inplace=True)
    df.fillna(0, inplace=True)
    return df

def add_revenue(df):
    df["Revenue"] = df["Quantity"] * df["Price"]
    return df

def main():
    df = load_data("sales.csv")
    df = clean_data(df)
    df = add_revenue(df)

    print("Total Revenue:", df["Revenue"].sum())
    print("Top Product:", df.groupby("Product")["Revenue"].sum().idxmax())

    summary = df.groupby("Product")["Revenue"].sum()
    summary.plot(kind="bar")
    plt.show()

if __name__ == "__main__":
    main()


```

A python based sales data analysis project that loads csv data,cleans it,calculates total revenue,identifies the top product,and visualizes results using bar.

efficient revenue analysis and visualization using python.
